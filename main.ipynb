```python
def save_uploaded_file(uploaded_file):
    """
    L∆∞u file ng∆∞·ªùi d√πng upload v√†o th∆∞ m·ª•c t·∫°m c·ªßa m√°y t√≠nh
    ƒë·ªÉ c√°c tool (OCR) c√≥ th·ªÉ ƒë·ªçc ƒë∆∞·ª£c ƒë∆∞·ªùng d·∫´n.
    """
    try:
        # T·∫°o file t·∫°m v·ªõi ƒëu√¥i m·ªü r·ªông g·ªëc (pdf, png, jpg...)
        suffix = "." + uploaded_file.name.split('.')[-1]
        with tempfile.NamedTemporaryFile(delete=False, suffix=suffix) as tmp_file:
            tmp_file.write(uploaded_file.getvalue())
            return tmp_file.name
    except Exception as e:
        st.error(f"L·ªói khi l∆∞u file: {e}")
        return None
```

```python
st.title("üïµÔ∏è‚Äç‚ôÇÔ∏è AI Resume & Career Analyzer")
st.markdown("### H·ªá th·ªëng ph√¢n t√≠ch CV, so kh·ªõp JD v√† g·ª£i √Ω l·ªô tr√¨nh s·ª± nghi·ªáp th√¥ng minh")
st.markdown("---")
```

```python
col1, col2 = st.columns(2)

```

```python
with col1:
    st.header("üìÑ Th√¥ng tin ·ª®ng vi√™n (CV)")
    
    # Ch·ªçn lo·∫°i ƒë·∫ßu v√†o
    cv_option = st.radio("Ch·ªçn ngu·ªìn d·ªØ li·ªáu CV:", ["Upload File (PDF)", "Nh·∫≠p vƒÉn b·∫£n (Text)"], key="cv_opt")
    
    cv_input = None
    cv_type = "file"
    
    if cv_option == "Upload File (PDF)":
        uploaded_cv = st.file_uploader("T·∫£i l√™n CV c·ªßa b·∫°n", type=["pdf"])
        if uploaded_cv:
            # L∆∞u file v√† l·∫•y ƒë∆∞·ªùng d·∫´n
            cv_input = save_uploaded_file(uploaded_cv)
            st.success(f"ƒê√£ t·∫£i: {uploaded_cv.name}")
    else:
        cv_type = "text"
        cv_input = st.text_area("D√°n n·ªôi dung CV v√†o ƒë√¢y:", height=300, placeholder="V√≠ d·ª•: Nguy·ªÖn VƒÉn A - Kinh nghi·ªám 3 nƒÉm Python...")
```

```python
with col2:
    st.header("üíº M√¥ t·∫£ C√¥ng vi·ªác (JD)")
    jd_option = st.radio("Ch·ªçn ngu·ªìn d·ªØ li·ªáu JD:", ["Upload File (PDF/·∫¢nh)", "Nh·∫≠p vƒÉn b·∫£n (Text)"], key="jd_opt")
    
    jd_input = None
    jd_type = "file"
    
    if jd_option == "Upload File (PDF/·∫¢nh)":
        uploaded_jd = st.file_uploader("T·∫£i l√™n JD", type=["pdf", "png", "jpg", "jpeg"])
        if uploaded_jd:
            jd_input = save_uploaded_file(uploaded_jd)
            st.success(f"ƒê√£ t·∫£i: {uploaded_jd.name}")
            if uploaded_jd.type.startswith('image'):
                st.image(uploaded_jd, caption="Preview JD", use_column_width=True)
                
    else:
        jd_type = "text"
        jd_input = st.text_area("D√°n n·ªôi dung JD v√†o ƒë√¢y:", height=300, placeholder="V√≠ d·ª•: C·∫ßn tuy·ªÉn Senior Python Developer, l∆∞∆°ng $2000...")
```

```python
st.markdown("---")
analyze_btn = st.button("üöÄ PH√ÇN T√çCH NGAY")
```

```python
if analyze_btn:
    if not cv_input or not jd_input:
        st.error("‚ö†Ô∏è Vui l√≤ng cung c·∫•p ƒë·∫ßy ƒë·ªß th√¥ng tin CV v√† JD tr∆∞·ªõc khi ph√¢n t√≠ch!")
    else:
        try:
            with st.spinner("ü§ñ AI ƒëang ƒë·ªçc t√†i li·ªáu, ph√¢n t√≠ch k·ªπ nƒÉng v√† t√¨m ki·∫øm kh√≥a h·ªçc... Vui l√≤ng ƒë·ª£i trong gi√¢y l√°t..."):
                result = analyze_cv_jd(
                    cv_input=cv_input,
                    jd_input=jd_input,
                    cv_type=cv_type,
                    jd_type=jd_type
                )
                st.success("‚úÖ Ph√¢n t√≠ch ho√†n t·∫•t!")
                
                st.markdown("## üìä B√°o C√°o Chi Ti·∫øt")
                st.markdown("---")
                st.markdown(result)
                
        except Exception as e:
            st.error(f"‚ùå C√≥ l·ªói x·∫£y ra trong qu√° tr√¨nh ph√¢n t√≠ch: {e}")
        finally:
            if cv_type == "file" and cv_input and os.path.exists(cv_input):
                os.unlink(cv_input)
            if jd_type == "file" and jd_input and os.path.exists(jd_input):
                os.unlink(jd_input)
```
